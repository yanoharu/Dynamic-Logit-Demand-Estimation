# 動学需要推定(J=1)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate,stats
import pandas as pd
from scipy.optimize import minimize
import scipy.special as sp
```

# Model

```python
class durable_goods_adoption_model():
    
    def __init__(self,
                ngrid = 30,
                n_min = 0.1,
                n_max = 10,
                nquad = 8,
                rho = [-0.01,0.9],
                sigma = 0.5,
                alpha = -1,
                beta = 5,
                delta = 0.95):
        
        self.ngrid = ngrid,
        self.grid = np.linspace(n_min,n_max,ngrid)
        self.nquad = nquad
        self.rho = rho
        self.sigma = sigma
        self.delta = delta
        self.alpha = alpha
        self.beta = beta
        self.quadrature_setup()
        
        
    def quadrature_setup(self):
        self.node, self.weight = np.polynomial.hermite.hermgauss(self.nquad)
        self.node = stats.norm(0,self.sigma).cdf(np.sqrt(2)*self.sigma*self.node)
        self.weight = self.weight/np.sqrt(np.pi) # array([w_1,...,w_n])
        
        
    def interp_func(self,x,f):
        """
        x is state gird.
        f is f(x) to be interpolated.
        
        """
        return interpolate.interp1d(x,f,kind='slinear',fill_value="extrapolate")
    
    
    def next_period_price(self,current_price):
        current_price = current_price[:,np.newaxis]
        next_price = self.rho[0] + self.rho[1]*current_price + self.node
        # next_price is shape(ngrid,nquad) array
        
        return next_price
    

    def bellman(self,V0):
        """
        bellman operator.
        V0 assume array([v_1,...v_ngrid])    
        """
        next_price = self.next_period_price(self.grid)
        inter = self.interp_func(self.grid,V0)
        V1 = inter(next_price)
        # V1 = array[[v_1_1,...,v_1_nquad],
        #            [v_2_1,...,v_2_nquad],
        #            [...................],
        #            [v_ngrid_1,...,v_ngrid_nquad]]
        
        V_purchase = self.beta+self.alpha*next_price     
        M = np.maximum(V1,V_purchase)
        EV1 = (np.log(np.exp(V1-M)+np.exp(V_purchase-M))+M)
        EV1 = self.delta*(EV1@self.weight[:,np.newaxis]).flatten()
        #EV1 = array([ev_1,...,ev_ngrid])
            
        current_V_purchase = self.beta+self.alpha*self.grid
        p0 = 1/(np.exp(EV1-EV1)+np.exp(current_V_purchase-EV1))
        p1 = 1-p0
        
        return EV1,V1,V0,V_purchase,p0,p1,next_price
    
    
    def solve_vfi (self,maxiter=1000000,tol=1e-10, callback=None):
        '''Solves the model using value function iterations (successive approximations of Bellman operator)
           Callback function is invoked at each iteration with keyword arguments.
        '''
        V0 = self.grid # initial guess
        for iter in range(maxiter):
            V1,V,v0,Vp,p0,p1,next_price = self.bellman(V0)
            err = np.amax(np.abs(V1-V0))
            if callback != None:callback(iter=i,err=err,V1=V1,V0=V0,p0=p0,p1=p1,model=self)
            if err < tol:
                break  # converged!
            V0 = V1  # prepare for the next iteration
        else:  # when iter went up to maxiter
            print(f"beta:{self.beta}")
            print(f"alpha:{self.alpha}")
            print(f"rho:{self.rho}")
            print(f"sigma:{self.sigma}")
            print(f"V1:{V1}")
            print(f"V0:{v0}")
            print(f"V:{V}")
            print(f"Vp:{Vp}")
            print(f"next_price:{next_price}")
            print(f"err:{err}")
            raise RuntimeError('No convergence: maximum number of iterations achieved!')
        return V1,p0,p1
    
    
    def solve_show(self,solver='vfi',maxiter=1000000,tol=1e-10,**kvargs):
        '''Illustrate solution'''
        fig1, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(14,8))
        ax1.set_xlabel('price grid')
        ax2.set_xlabel('price grid')
        ax3.set_xlabel('price grid')
        ax1.set_title('Value function')
        ax2.set_title('Probability of outside option')
        ax3.set_title('Probability of buying')
        def callback(**argvars):
            model,V1,p0,p1=argvars['model'],argvars['V1'],argvars['p0'],argvars['p1']
            ax1.plot(self.grid,V1,color='k',linewidth=0.25)
            ax2.plot(self.grid,p0,color='k',linewidth=0.25)
            ax3.plot(self.grid,p1,color='k',linewidth=0.25)
        # add solutions
        V1,p0,p1 = self.solve_vfi(maxiter=maxiter,tol=tol)
        ax1.plot(self.grid,V1,color='r',linewidth=2.5)
        ax2.plot(self.grid,p0,color='r',linewidth=2.5)
        ax3.plot(self.grid,p1,color='r',linewidth=2.5)
        plt.show()
        return V1,p0,p1
        

model = durable_goods_adoption_model()
V1,p0,p1 = model.solve_show()
```

## Price process simulation

```python
def simulate_price_process(T = 30,
                           price_0 = 7,
                           sigma = model.sigma,
                           rho = model.rho):
    """
    Return price process
    First column is price process for product 1 and second for product 2.
    Number of row corresponds to time period
    """
    np.random.seed(1234)
    psi = np.random.normal(0, sigma, size=T)
    price = np.zeros(T)
    price[0] = np.array(price_0)
    for t in range(1,T):
        price[t] = rho[1]*price[t-1] + rho[0] + psi[t]
    
    return price

plt.plot(simulate_price_process())
```

## Data simulation

- 初期マーケットサイズをN_consumerで指定し財を購入した消費者はマーケットから退出し、購入を控えた消費者はマーケットに残ります。

```python
def simulate_purchase_decision(T = 100,
                               N_consumer = 100000,
                               price_0 = 9,
                               model = model,
                               sigma = model.sigma,
                               rho = model.rho):
    
    
    beta = model.beta
    alpha = model.alpha
    price = simulate_price_process(T,price_0,sigma=sigma,rho=rho)
    V0,_,_ = model.solve_show()
    inter = model.interp_func(model.grid,V0)
    data_array = np.array([[],[],[],[]])
    # first dim store idx of consumer, second dim for time
    #third dim for decision, forth dim for price

    
    for i in range(N_consumer):
        idx_consumer = []
        time = []
        decision_i = []
        price_i = []
        for t in range(T):
            idx_consumer.append(i)
            time.append(t)
            np.random.seed(i+t)
            epsilon = sp.digamma(1)-np.log(-np.log(np.random.rand(2)))        
            v0 = inter(price[t]) + epsilon[0]
            v1 = beta + alpha*price[t]+ epsilon[1]
            decision = 1 if v1  > v0 else 0
            decision_i.append(decision)
            price_i.append(price[t])
            if decision != 0:
                  break
        data_i = np.array([idx_consumer,time,decision_i,price_i])
        data_array = np.hstack((data_array,data_i))
    
    data = pd.DataFrame(np.transpose(data_array),columns=["i","t","decision","state"])
    
    plt.plot(data["decision"].groupby(data["t"]).sum(),label = "Sales")
    plt.legend()
    plt.show()
    plt.plot(simulate_price_process(T=len(np.unique(data["t"])),price_0=price_0,sigma=model.sigma,rho=model.rho),label ="price")
    plt.legend()
    plt.show()
    
    pre_state = (data.state.groupby(data.t).unique().shift(1)).apply(lambda x: float(str(x[0])) if pd.notna(x) else None)
    data = pd.merge(data, pd.concat([pre_state], axis=1), how ='left', on='t', suffixes=( '','_previous'))
    
    
    return data
data = simulate_purchase_decision()
data
```

## Estimation

```python
def log_likelihood(theta = [2,-0.9,-0.03,1,0.5],data=data):
    model = durable_goods_adoption_model(beta = theta[0],alpha = theta[1],rho=theta[2:4],sigma=theta[4])
    V,_,_ = model.solve_vfi()
    beta = theta[0]
    alpha = theta[1]
    rho = theta[2:4]
    sigma = theta[4]
    T = len(data.t.unique())
    inter = model.interp_func(model.grid,V)

    Log_lik = 0
    
    data = data[data.t != 0]
    
    # no purchase
    V0 = inter(data.state[data.decision==0])
    V_purchase = beta + alpha*data.state[data.decision==0]
    p0 = 1/(np.exp(V0-V0)+np.exp(V_purchase-V0))   
    state_pred = rho[0] + rho[1]*data.state_previous[data.decision==0]
    psi = data.state[data.decision==0] - state_pred
    f_psi = stats.norm(0,sigma).pdf(psi)
              
    Log_lik += np.sum(np.log(p0)+ np.log(f_psi))
    
    # purchase good 
    V0 = inter(data.state[data.decision==1])
    V_purchase = beta + alpha*data.state[data.decision==1]
    p0 = 1/(np.exp(V0-V0)+np.exp(V_purchase-V0))   
    p1 = 1 - p0
    state_pred = rho[0] + rho[1]*data.state_previous[data.decision==1]
    psi = data.state[data.decision==1] - state_pred
    f_psi = stats.norm(0,sigma).pdf(psi)
              
    Log_lik += np.sum(np.log(p1)+ np.log(f_psi))
             
    
    return -Log_lik

log_likelihood(theta = [model.beta,model.alpha,model.rho[0],model.rho[1],model.sigma],data=data)
```

```python
%%time
bounds = [(-100,100),(-100,0),(-1,1),(0,1),(0,1)]
theta_init = (3,-3,0.3,0.3,0.3)
θ_star = minimize(log_likelihood, x0=theta_init, args=(data),method='L-BFGS-B',bounds=bounds).x
θ_star
```
